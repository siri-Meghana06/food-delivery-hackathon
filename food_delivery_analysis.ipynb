```python
# import libraries
import pandas as pd
import sqlite3
```

```python
# Load orders.csv
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
# Load users.csv
users = pd.read_json("users.json")
users.head()
```

```python
# Load restaurants.sql
conn = sqlite3.connect(":memory:")

with open("restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
#orders+users
df = orders.merge(users, on="user_id", how="left")
```

```python
# +restaurants
df = df.merge(restaurants, on="restaurant_id", how="left")
df.head()
```

## Dataset Merging
- orders.csv joined with users.json using user_id
- restaurants.sql joined using restaurant_id
- Left join used to retain all orders
